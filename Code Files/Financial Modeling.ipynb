```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df_hp = pd.read_excel('Health Insurance Premium.xlsx')
```

```python
df_hp.head()
```

```python
df_hp.dtypes
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.figure(figsize=(6, 6))
sns.barplot(x='Financial Year', y='Gross Premium Collection of Health Insurance', data=df_hp)

plt.xlabel('Financial Year')
plt.ylabel('Gross Premium Collection of Health Insurance in INR Cr')
plt.title('Gross Premium Collection of Health Insurance in each Financial Year')
plt.savefig('Premium v FY .png', dpi=300, bbox_inches='tight')
```

```python
df_nh['Year'] = df_nh['Financial Year'].str[-2:].astype(int)
```

```python
df_nh['Year'] = df_nh['Year']+2000
```

```python
df_nh.head()
```

```python
X = df_nh[['Year']]
y=  df_nh['Total number of people having health insurance']
```

```python
X
```

```python
y
```

```python
# Create an instance of the LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict using the trained model
y_pred = model.predict(X)

# Print the coefficients and intercept of the linear regression model
print('Coefficient:', model.coef_[0])
print('Intercept:', model.intercept_)
```
